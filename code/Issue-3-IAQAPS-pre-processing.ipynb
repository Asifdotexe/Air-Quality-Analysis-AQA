When we have missing values, we have it for the entire row of data, likey due to sensor malfunctions, we can impute this data with mean to deal with them.

```python
# checking for all the unique pollutants in the dataset
df.pollutant_id.unique()
```

```python
# List of all possible pollutants
possible_pollutants = ['NH3', 'SO2', 'PM2.5', 'PM10', 'OZONE', 'CO', 'NO2']

def impute_missing_values(df, possible_pollutants):
    # Group by 'pollutant_id' and calculate the mean pollutant values for each group
    pollutant_means = df.groupby('pollutant_id')[['pollutant_min', 'pollutant_avg', 'pollutant_max']].mean().to_dict(orient='index')
    
    # Function to impute missing values based on pollutant category means
    def impute_missing(row):
        pollutant_id = row['pollutant_id']
        if pd.isnull(row['pollutant_min']):
            row['pollutant_min'] = pollutant_means.get(pollutant_id, {}).get('pollutant_min', None)
        if pd.isnull(row['pollutant_avg']):
            row['pollutant_avg'] = pollutant_means.get(pollutant_id, {}).get('pollutant_avg', None)
        if pd.isnull(row['pollutant_max']):
            row['pollutant_max'] = pollutant_means.get(pollutant_id, {}).get('pollutant_max', None)
        return row
    
    # Apply the imputation function to each row of the DataFrame
    df_imputed = df.apply(impute_missing, axis=1)
    
    return df_imputed
```

```python
df_imputed = impute_missing_values(df, possible_pollutants)
```
